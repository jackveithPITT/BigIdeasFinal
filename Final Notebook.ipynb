## Analysis of Crime Statistics in Pittsburgh Neighborhoods
In order to determine the best, most family-friendly neighborhood in Pittsburgh, I (Jack V) chose to analyze the quantity of different types of crimes committed in each neighborhood, and how that compared to the population of each neighborhood.

```python

#read in the csv as data
data = pd.read_csv("Neighborhoods_with_SNAP_Data.csv")

#these are the names of the colummns we are interested in
col_names = ['SNAP_All_csv__Murder__2010_',
             'SNAP_All_csv__Rape__2010_',
             'SNAP_All_csv__Burglary__2010_',
             'SNAP_All_csv__Auto_Theft__2010_',
             'SNAP_All_csv__Robbery__2010_',
             'SNAP_All_csv__Drug_Violations__'
             ]

#not all crimes are equal so ive created a naive way to weight the
#types of crime committed
crime_weights = [100,
                 50,
                 50,
                 10,
                 10,
                 5
                ]
    
#set containing each neighborhood and the corresponding real values
#of each type of crime committed in the area
nbhd_set = {}

#init the set
for name in data['Neighborhood_2010_HOOD']:
    nbhd_set[name] = (name, [0] * 6)


#for each type of crime
for i in range(6):
    for x in range(len(data['Neighborhood_2010_HOOD'])):
        #iterate over the csv and fill the nbhd_set with the counts of each crime committed
        nbhd_set[data['Neighborhood_2010_HOOD'][x]][1][i] += data[col_names[i]][x]
        

#set containing the weighted total representing the crime in each neighborhood
aggregate_set = {}


#for each neighborhood
for nbhd in nbhd_set:
    #initialize the set
    aggregate_set[nbhd_set[nbhd][0]] = 0
    #for each type of crime
    for i in range(6):
        #add the weighted score of each crime to the aggregate_set
        aggregate_set[nbhd_set[nbhd][0]] += crime_weights[i] * nbhd_set[nbhd][1][i]
        
sorted_aggregate = dict(sorted(aggregate_set.items(),key= lambda x:x[1]))

df1 = pd.DataFrame({'Neighborhood':list(sorted_aggregate.keys()), 'Weighted Crime Score':list(sorted_aggregate.values())})
df1.plot.bar(x="Neighborhood", y = "Weighted Crime Score", figsize=(20, 10))
```

### Halfway Conclusion
the above code is a bit messy, but is explained somewhat briefly within the comments. To explain in plain english, essentially what I did was create a dictionary that holds a tuple for each neighborhood containing two pieces of information: the name of the neighborhood, and an array that contains the counts of each type of crime committed. I also created a naive weight for each type of crime, as murder and burglary should not, in my opinion, hold the same weight. I iterated over the initial dictionary, creating a new dictionary that contains an integer value which represents the sum of all crimes in the neighborhood multiplied by the weight of the crimes.

### One Small Issue
it is obvious that we cant draw a conclusion simply from this data; each neighborhood has a different quantity of citizens, and thus total counts of crimes will be higher in certain areas simply due to the large quantity of people. Therefore I had to augment the data found previously by dividing each neighborhood data point by the population of the neighborhood.

```python
        
#pop_set is the population of each neighborhood which we retrieve here from the csv (2010)
pop_set = [0] * len(data['Pop__2010'])
for i in range(len(data['Pop__2010'])):
    pop_set[i] = data['Pop__2010'][i]

#simply dividing the aggregate weighted scores of the crimes by the population of each neighborhood
iterator = 0
for nbhd in nbhd_set:
    aggregate_set[nbhd_set[nbhd][0]] = aggregate_set[nbhd_set[nbhd][0]] / pop_set[iterator]
    iterator+=1

    
sorted_aggregate_pop = dict(sorted(aggregate_set.items(),key= lambda x:x[1]))
    
#list(sorted_aggregate.keys())

#aggregate_set.keys()

df2 = pd.DataFrame({'Neighborhood':list(sorted_aggregate_pop.keys()), 'Weighted Crime Score':list(sorted_aggregate_pop.values())})
g = df2.head(15)
g.plot.bar(x="Neighborhood", y = "Weighted Crime Score", figsize=(20,10), rot=45)
```
